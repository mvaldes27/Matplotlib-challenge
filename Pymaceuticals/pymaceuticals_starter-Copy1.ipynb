## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

#make dataframe
mouse_data_df = pd.DataFrame(mouse_metadata)
study_results_df = pd.DataFrame(study_results)





# Combine the data into a single dataset

mice_merged_df = pd.merge(mouse_data_df, study_results_df, on="Mouse ID", how="left")

mice_merged_df





# Display the data table for preview
```

## Summary Statistics

```python
#define function that returns SEM

def sem(array):
        
    return np.std(array) / np.sqrt(np.size(array))



    
```

```python
#agg statistical functions by Drug Regimen
drug_stats_summary = mice_merged_df_clean.groupby(['Drug Regimen'])['Tumor Volume (mm3)'].agg([np.mean,np.std, np.median, np.var, sem])
# mice_merged_df.groupby([])[].describe()
# mice_merged_df.groupby([])[].mean()

drug_stats_summary
```

```python
#compare the created function to scipy equivalent

mice_merged_df.groupby(['Drug Regimen'])['Tumor Volume (mm3)'].agg([st.sem])

```
